```python
import plotly.express as px
from datetime import datetime
import pytz

# ------------------------------
# Time Restriction (3PMâ€“5PM IST)
# ------------------------------
ist = pytz.timezone("Asia/Kolkata")
current_time = datetime.now(ist)

if 15 <= current_time.hour < 17 and not top_categories.empty:

    # Reset index for Plotly
    plot_df = top_categories.reset_index()

    # Convert to long format for grouped bar
    plot_df = plot_df.melt(
        id_vars="Category",
        value_vars=["Rating", "Reviews"],
        var_name="Metric",
        value_name="Value"
    )

    # Create grouped bar chart
    fig = px.bar(
        plot_df,
        x="Category",
        y="Value",
        color="Metric",
        barmode="group",
        title="Average Rating vs Total Reviews (Top 10 Categories by Installs)",
        labels={
            "Category": "App Category",
            "Value": "Values"
        }
    )

    fig.update_layout(xaxis_tickangle=-45)

    fig.show()
    import plotly.io as pio
    pio.write_html(fig, "task_1.html", auto_open=False)

else:
    print("This graph is visible only between 3 PM IST and 5 PM IST.")
```
